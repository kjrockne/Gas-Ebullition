```python
#Libraries needed to run the tool
import numpy as np
import pandas as pd
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
import xgboost
from xgboost import XGBRegressor
```

```python
#Ask for file name and read the file
file_name = 'Master_Compiled'
df = pd.read_csv(file_name + '.csv', header=0, index_col=0)
```

```python
#Data preprocessing
df= df.drop(columns=["Date", 'Temp_K','Gas Production Rate (mmol/m3/d)', 'Moisture' ])

df.dropna(inplace=True)
#remove NAN
df.dropna(subset=['COD/TOC_Molar'], inplace=True)
index_names1 = df[ df[ 'COD/TOC_Molar'] == "" ].index
df.drop(index_names1, inplace = True)
index_names1 = df[ df[ 'depth'] == "" ].index
df.drop(index_names1, inplace = True)
index_names1 = df[ df[ 'Temp_C'] == "" ].index
df.drop(index_names1, inplace = True)
index_names1 = df[ df[ 'Solids'] == "" ].index
df.drop(index_names1, inplace = True)
index_names1 = df[ df[ 'Porosity'] == "" ].index
df.drop(index_names1, inplace = True)
index_names1 = df[ df[ 'Bulk_Density'] == "" ].index
df.drop(index_names1, inplace = True)
index_names1 = df[ df[ 'OM'] == "" ].index
df.drop(index_names1, inplace = True)
index_names1 = df[ df[ 'OC'] == "" ].index
df.drop(index_names1, inplace = True)
index_names1 = df[ df[ 'COD'] == "" ].index
df.drop(index_names1, inplace = True)
index_names1 = df[ df[ 'OC/OM'] == "" ].index
df.drop(index_names1, inplace = True)
data = df
print (data)
#Print number of rows and colums read
print("{0} rows and {1} columns".format(len(data.index), len(data.columns)))
print("")
```

```python
#Defining X and Y
X = data[['depth', 'Temp_C', 'Solids','Porosity','Bulk_Density', 'OM', 'OC',
'COD','OC/OM', 'COD/TOC_Molar']]
Y = data.Gasflux_mM_m2d
```

```python
np.random.seed(8)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state=42)
```

```python
# Scaling the data
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_scaled = scaler.transform(X)
```

```python
#pip install scikit-learn==1.2.2
```

```python
xgb = XGBRegressor(colsample_bytree= 1, learning_rate= 0.01, max_depth= 2, reg_alpha=1,
  n_estimators= 200, subsample= 1, random_state=42)
xgb.fit(X_train_scaled, Y_train)
```
